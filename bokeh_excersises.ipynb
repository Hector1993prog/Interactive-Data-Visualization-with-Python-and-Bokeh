## 2.  Customizing Bokeh graphs

### 2.1. Creating an initial plot with the iris dataset

```python
from bokeh.io import output_notebook 
'''
output_notebook = Configure the default output state to generate output in notebook cells when |show| is called. 
Note that |show| may be called multiple times in a single cell 
to display multiple objects in the output cell. The objects will be displayed in order.
'''
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
flor_plot= Figure()

colormap={'setosa':'red', 'versicolor':'green', 'virginica':'blue'}
flowers['color']=[colormap[x] for x in flowers['species']]

flor_plot.circle(x= flowers['petal_length'], y= flowers['petal_width'], size=flowers['sepal_width']*4,
fill_alpha= 0.2, color= flowers.color)
output_file('flor_plot.html')
show(flor_plot)
```

### 2.2. Background and title

#### 2.2.1. Creating Backgrounds

```python
flor_plot.plot_width=1100
flor_plot.plot_height=650
flor_plot.background_fill_color= 'olive'
flor_plot.background_fill_alpha=0.3
output_file('first customized plot.html')
show(flor_plot)
```

#### 2.2.2 Creating the Tittle: Size & fonts

```python
output_file('second customized plot.html')
flor_plot.title.text='Iris Morphology'
flor_plot.title.text_color= 'olive'
flor_plot.title.text_font='times'
flor_plot.title.text_font_size="25px"
flor_plot.title.align= 'center'
show(flor_plot)
```

### 2.3. Axes

#### 2.3.1 Custom styling

```python
output_file('4. customizing the axis.html')
flor_plot.xaxis.minor_tick_in= -6
flor_plot.yaxis.major_label_orientation='vertical'
flor_plot.xaxis.minor_tick_line_color='blue'
flor_plot.xaxis.visible=True
flor_plot.xaxis.axis_label= 'Petal Length'
flor_plot.yaxis.axis_label= 'Petal width'
flor_plot.axis.axis_label_text_color='blue'
flor_plot.axis.major_label_text_color='orange'
show(flor_plot)
```

#### 2.3.2 Axes: Custom geometry

```python
from bokeh.models import Range1d
output_file('5. custom geometry of the axis.html')
flor_plot.x_range= Range1d(start=0, end=10) #range of the axis
flor_plot.y_range=Range1d(0,5)
flor_plot.xaxis.bounds=(2, 6)# limits to the panning method of the plot
flor_plot.xaxis[0].ticker.desired_num_ticks= 2# tickers in the axes. Yo always has to set a position with []
flor_plot.yaxis[0].ticker.desired_num_ticks=2
flor_plot.yaxis[0].ticker.num_minor_ticks=10
show(flor_plot)


```

### 2.4. Axes: Categorical data

```python
categorical= Figure(x_range=["F", "D-", "D", "D+", "C-", "C", "C+", "B-", "B", "B+", "A-", "A", "A+"],
y_range=["F", "D-", "D", "D+", "C-", "C", "C+", "B-", "B", "B+", "A-", "A", "A+"])


categorical.circle(x=["A","B"], y=["C","D"], size=8)


output_file('6. categorical data in the axis.html')

show(categorical)
```

### 2.5. working the grid

```python
output_file('customizing grid.html')
flor_plot.xgrid.grid_line_color= None
flor_plot.ygrid.grid_line_alpha=0.6
flor_plot.grid.grid_line_dash=[3,5]
show(flor_plot)
```

### 2.6. customizing the tool bar of bokeh plots

```python
from bokeh.models import PanTool, ResetTool, HoverTool
flor_plot.tools= [PanTool(), ResetTool()]# selectiong the tools you want to show in the plot
flor_plot.add_tools(HoverTool()) #adding more tools to your list
flor_plot.toolbar_location = 'above' #position of the tools
flor_plot.toolbar.logo = None # 
output_file('8. custom tools.html')
show(flor_plot)
```

### 2.8. Legend 

#### 2.8.1. Configuration

```python
output_file('10. configuring the lenyend.html')
# se añaden tres sets para cada uno de los puntos. En este caso se usa la 
# sintaxis de pandas para especificar cada uno

flor_plot.circle(x=flowers["petal_length"][flowers["species"]=="setosa"], 
y=flowers["petal_width"][flowers["species"]=="setosa"], 
size=flowers['sepal_width'][flowers["species"]=="setosa"]*4,
fill_alpha=0.2,color=flowers['color'][flowers["species"]=="setosa"],legend_label='Setosa')

flor_plot.circle(x=flowers["petal_length"][flowers["species"]=="versicolor"], 
y=flowers["petal_width"][flowers["species"]=="versicolor"],
size=flowers['sepal_width'][flowers["species"]=="versicolor"]*4,
fill_alpha=0.2,color=flowers['color'][flowers["species"]=="versicolor"],legend_label='Versicolor')

flor_plot.circle(x=flowers["petal_length"][flowers["species"]=="virginica"], 
y=flowers["petal_width"][flowers["species"]=="virginica"],
size=flowers['sepal_width'][flowers["species"]=="virginica"]*4,
fill_alpha=0.2,color=flowers['color'][flowers["species"]=="virginica"],legend_label='Virginica')

show(flor_plot)
```

```python
#Plotting flower species
 
#Importing libraries
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.sampledata.iris import flowers
from bokeh.models import Range1d, PanTool, ResetTool, HoverTool, WheelZoomTool
from screeninfo import get_monitors
 
#Define the output file path
output_file("iris.html")
 
#Create the figure object
f = figure(tools=[PanTool(), ResetTool(), WheelZoomTool()])
 
#Style the tools
hover = HoverTool(tooltips=[("Species","@species"), ("Sepal Width", "@sepal_width")])
f.add_tools(hover)
f.toolbar_location = 'above'
f.toolbar.logo = None
 
#Style the plot area
f.plot_width=get_monitors()[0].width #get_monitors is part of the screeninfo module imported above
f.plot_height=get_monitors()[0].height-50 #get_monitors is part of the screeninfo module imported above
f.background_fill_color = "grey"
f.background_fill_alpha = 0.1
f.sizing_mode = "stretch_both" #graph will resize itself when user resizes the browser
 
#Style the title
f.title.text = "Iris Morphology"
f.title.text_color = "olive"
f.title.text_alpha = 0.6
f.title.text_font = "antiqua"
f.title.text_font_size = "18px"
f.title.align = "center"
 
#Style the axes
f.yaxis.major_label_orientation = "vertical"
f.xaxis.minor_tick_in = -6
f.yaxis.minor_tick_in = -6
f.axis.minor_tick_line_color = "grey"
f.axis.axis_line_color = "olive"
f.xaxis.axis_label = "Petal Length"
f.yaxis.axis_label = "Petal Width"
f.axis.axis_label_text_color = "olive"
f.axis.axis_label_text_font = "antiqua"
f.axis.axis_label_text_font_style = "bold"
f.axis.major_label_text_color = "olive"
 
#Axes geometry
f.x_range = Range1d(start=flowers["petal_length"][flowers["species"]=='versicolor'].min(),
end=flowers["petal_length"][flowers["species"]=='versicolor'].max())
f.y_range = Range1d(start=flowers["petal_width"][flowers["species"]=='versicolor'].min(),
end=flowers["petal_width"][flowers["species"]=='versicolor'].max())
f.xaxis[0].ticker.desired_num_ticks = 6
f.yaxis[0].ticker.desired_num_ticks = 6
f.yaxis[0].ticker.num_minor_ticks = 5
f.xaxis[0].ticker.num_minor_ticks = 5
 
#Style the grid
f.grid.grid_line_color = (128,128,0,0.5) #equivalent to olive color with a 0.5 alpha value
f.grid.grid_line_dash = [5,3] #5 pixels of line and 3 pixels of space
f.grid.minor_grid_line_color = (128,128,0,0.1) #equivalent to olive color with a 0.1 alpha value
f.grid.minor_grid_line_dash = [3,3]
 
colormap = {'setosa':'red','versicolor':'green','virginica':'blue'}
flowers['color'] = [colormap[x] for x in flowers['species']]
 
#adding glyphs
f.circle(x=flowers["petal_length"][flowers["species"]=="setosa"],
y=flowers["petal_width"][flowers["species"]=="setosa"],
size=flowers['sepal_width'][flowers["species"]=="setosa"]*4,
fill_alpha=0.2, color=flowers['color'][flowers["species"]=="setosa"],
legend_label='Setosa')
 
f.circle(x=flowers["petal_length"][flowers["species"]=="versicolor"],
y=flowers["petal_width"][flowers["species"]=="versicolor"],
size=flowers['sepal_width'][flowers["species"]=="versicolor"]*4,
fill_alpha=0.2, color=flowers['color'][flowers["species"]=="versicolor"],
legend_label='Versicolor')
 
f.circle(x=flowers["petal_length"][flowers["species"]=="virginica"],
y=flowers["petal_width"][flowers["species"]=="virginica"],
size=flowers['sepal_width'][flowers["species"]=="virginica"]*4,
fill_alpha=0.2, color=flowers['color'][flowers["species"]=="virginica"],
legend_label='Virginica')
 
#Style the legend
f.legend.location = "top_left"
f.legend.background_fill_alpha = 0
f.legend.border_line_color = None
f.legend.margin = 10
f.legend.padding = 18
f.legend.spacing = 1
f.legend.label_text_color = 'olive'
f.legend.label_text_font = 'antiqua'
 
#Save and show the figure
show(f)
 
```
